Description: 
You are given two non-empty linked lists representing two non-negative integers. The digits are stored in reverse order, and each of their nodes contains a single digit. Add the two numbers and return the sum as a linked list.

You may assume the two numbers do not contain any leading zero, except the number 0 itself.

Solution to: 2. Add Two Numbers
URL: https://leetcode.com/problems/add-two-numbers/

```python
from typing import List
from typing import Optional
# Definition for singly-linked list.
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

class Solution:
    def addTwoNumbers(self, l1: Optional[ListNode], l2: Optional[ListNode]) -> Optional[ListNode]:
            lista = []
            ext = 0
            while l1 != None or l2 != None:
                if l1 != None:
                    x = l1.val
                else:
                    x = 0
                if l2 != None:
                    y = l2.val
                else:
                    y = 0
                sum = x + y
                if ext != 0:
                    sum = ext + sum
                    ext = 0
                if sum >= 10:
                    ext = (sum) - 10 
                    lista.append(ext)
                    ext = 1
                else:
                    lista.append(sum)
                if l1 is not None:
                    l1 = l1.next
                if l2 is not None:
                    l2 = l2.next
            if ext != 0:
                lista.append(ext)
            node = ListNode(lista[0])
            l1 = node
            for n in range(1,len(lista)):
                newnode = ListNode(lista[n])
                node.next = newnode
                node = newnode
            return l1
        


if __name__ == "__main__":
    lista1 = [9,9,9,9,9,9,9]
    node = ListNode(lista1[0])
    l1 = node
    for n in range(1,len(lista1)):
        newnode = ListNode(lista1[n])
        node.next = newnode
        node = newnode
    
    lista2 = [9,9,9,9]
    node = ListNode(lista2[0])
    l2 = node
    for n in range(1,len(lista2)):
        newnode = ListNode(lista2[n])
        node.next = newnode
        node = newnode
        
    solution = Solution()
    result = solution.addTwoNumbers(l1, l2)
    print(result)

```
